```python
# Now lets load the 2nd csv file
df1 = pd.read_csv('CountryCode.csv')
df1
```

```python
# Lets merge this to dataset
data_1 = pd.merge(df,df1, on = 'Country Code' , how = 'left')
data_1.head()
```

```python
#Data Analysis

data_1.nunique()
     
```

```python
# Now lets check how many outlets are on Zomato
data_1['Restaurant Name'].value_counts()
     
```

```python

# Lets see countrywise distribution
country_zomato = pd.DataFrame(data_1['Country'].value_counts())
country_zomato
```

```python
# Reset index to make 'Country' a column
country_zomato.reset_index(inplace=True)

# Add the percentage column
country_zomato['%'] = round(country_zomato['count'] * 100 / len(data_1), 2)

# Print the updated DataFrame
print(country_zomato.head())
```

```python
# Lets combine small values
top_2 = country_zomato.iloc[0:2]  # Use .iloc instead of [] for positional indexing
others = pd.DataFrame(country_zomato.iloc[2:].sum())  # Use .iloc instead of [] for positional indexing
others = others.T
others = others.rename(index={0: 'others'})
country = pd.concat([top_2, others], ignore_index=True)
country
```

```python
# Extract the count and percentage values from the 'others' DataFrame
others_count = others['count'].values[0]
others_percentage = others['%'].values[0]

# Create a new row for 'others' with the combined count and percentage
others_row = pd.DataFrame({'Country': ['Others'], 'count': [others_count], '%': [others_percentage]})

# Concatenate 'top_2' with the new 'others' row
country = pd.concat([top_2, others_row], ignore_index=True)
country
```

```python
# Plot of top 5 country who has most of the restaurants
plt.figure(figsize = (12,6))
explode = [0,0.3,0.3]
plt.pie(country['%'],labels = country.index,autopct='%1.2f%%',explode = explode)
plt.title("Pie chart of use of Zomato", fontsize = 15)
plt.show()
     
```

```python
# Lets check top 10 cuisine
cuisines = pd.DataFrame(data_1['Cuisines'].value_counts()).head(10)
cuisines = cuisines.reset_index().rename(columns = {'index':'Cuisines'})
cuisines
     
```

```python
plt.figure(figsize=(13, 6))
plots = sns.barplot(data=cuisines, x='Cuisines', y='count') 

for bar in plots.patches:
    plots.annotate(format(bar.get_height()),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=12, xytext=(0, 8),
                   textcoords='offset points')

plt.title('Top 10 Cuisines', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
# What type of currencies accepted by zomato
currency = pd.DataFrame(data_1.groupby(['Country'])['Currency'].value_counts())
currency = currency.rename(columns = {'Currency':'Count'})
currency
```

```python
# How many reastaurants have online delivery services
online_delivery = data_1.groupby('Country')['Has Online delivery'].value_counts()
online_delivery
```

```python
#Only India and UAE has online delivery services rest are not

# Distribution of price range
plots = sns.countplot(x = data_1['Price range'])
for bar in plots.patches:
  plots.annotate(format(bar.get_height()),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=12, xytext=(0, 8),
                   textcoords='offset points')
plt.title('Countplot of price range')
plt.ylim(0,5000)
plt.show()
```

```python
# Aggregate rating distribution
plt.figure(figsize = (15,6))
sns.countplot(x = data_1['Aggregate rating'])
plt.title('Countplot of Aggregate rating')
plt.show()
```

```python
# Lets get ratings dataframe 
ratings = data_1.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
# There are lot of Not rated restaurants 
data_1[data_1['Rating text'] == 'Not rated'].groupby(['Country']).size()
```

```python
## Count plot
plt.figure(figsize = (15,6))
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
plt.show()
```
